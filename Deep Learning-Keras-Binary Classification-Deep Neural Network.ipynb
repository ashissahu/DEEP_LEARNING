```python
np.random.seed(0)
n_pts = 500
X, y = datasets.make_circles(n_samples=n_pts, random_state = 123, noise=0.1, factor=0.2)
#factor=0.2 specifies that diameter ofinner circle is 20% ofouter circle diameter
#we are addinganoise =0.1,to make the modelmoredifficult topredict
#print(X) 
#print(y)
 
plt.scatter(X[y==0, 0], X[y==0, 1]) #plotting all the points with y=0=label i.eforouter circle points
plt.scatter(X[y==1, 0], X[y==1, 1]) ##plotting all the points with y=1=label i.e. for inner circle points
```

```python
model = Sequential()
model.add(Dense(4, input_shape=(2,), activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
model.compile(Adam(lr = 0.01), 'binary_crossentropy', metrics=['accuracy'])
h=model.fit(x=X, y=y, verbose=1, batch_size=50,epochs=500, shuffle='true')
```

```python
plt.plot(h.history['accuracy'])
plt.legend(['accuracy'])
plt.title('accuracy')
plt.xlabel('epoch')
 
```

```python
plt.plot(h.history['loss'])
plt.legend(['loss'])
plt.title('loss')
plt.xlabel('epoch')
```

```python
def plot_decision_boundary(X, y, model):
    x_span = np.linspace(min(X[:,0]) - 0.25, max(X[:,0]) + 0.25)
    y_span = np.linspace(min(X[:,1]) - 0.25, max(X[:,1]) + 0.25)
    xx, yy = np.meshgrid(x_span, y_span)
    grid = np.c_[xx.ravel(), yy.ravel()]
    pred_func = model.predict(grid)
    z = pred_func.reshape(xx.shape)
    plt.contourf(xx, yy, z)


```

```python
plot_decision_boundary(X, y, model)
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
```

# Predicting for a random point 

```python
#case 1
#==========

plot_decision_boundary(X, y, model)
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
 
x = 0.1
y = 0.75
 
point = np.array([[x, y]])
predict = model.predict(point) #function used to predict value 
plt.plot([x], [y], marker='o', markersize=10, color="red")
print("Prediction is: ", predict)


#If we see the ouptput, it indicates , a very low probabality(0.00094711. i.e0.094%) of 1. SO It belongs to class 0
```
